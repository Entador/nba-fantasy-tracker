```python
# Imports
import time
import pandas as pd
from nba_api.stats.static import players
from nba_api.stats.endpoints import (
    playercareerstats,
    playergamelog,
)

# -----------------------------
# CONFIG
# -----------------------------
PLAYER_NAME = "Stephen Curry"
SEASON = "2025-26"

# -----------------------------
# Get Player ID
# -----------------------------
player = players.find_players_by_full_name(PLAYER_NAME)[0]
player_id = player["id"]
print(f"{PLAYER_NAME} Player ID:", player_id)

# -----------------------------
# Career Season Totals
# -----------------------------
career = playercareerstats.PlayerCareerStats(player_id=1631128)
career_df = career.get_data_frames()[0]

season_totals = career_df[career_df["SEASON_ID"] == SEASON]
print("\nSeason Totals:")
display(season_totals)

# -----------------------------
# Game-by-Game Regular Season
# -----------------------------
game_log_regular = playergamelog.PlayerGameLog(
    player_id=player_id,
    season=SEASON,
    season_type_all_star="Regular Season"
)
regular_df = game_log_regular.get_data_frames()[0]

time.sleep(1)

# -----------------------------
# Game-by-Game Playoffs
# -----------------------------
game_log_playoffs = playergamelog.PlayerGameLog(
    player_id=player_id,
    season=SEASON,
    season_type_all_star="Playoffs"
)
playoffs_df = game_log_playoffs.get_data_frames()[0]

# -----------------------------
# Combine All Games
# -----------------------------
all_games_df = pd.concat([regular_df, playoffs_df], ignore_index=True)
print("\nAll Games:")
display(all_games_df.head())
```
